```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from plotly.subplots import make_subplots
import plotly.graph_objects as go

coinglass_api_key = '91468f8a11bc4f2587cb250789da75dd'
```

```python
def download_pair(url: str, start_time: datetime, ex: str, pair: str, interval: str) -> pd.DataFrame:
    
    data = []

    headers = {
               "accept": "application/json",
               "coinglassSecret": "91468f8a11bc4f2587cb250789da75dd"
              }
    
    while start_time <= datetime.now():
        
        # Set the parameters for the API request
        params = {
                  "ex": ex,
                  "pair": pair,
                  "interval": interval,
                  "limit": 5000,
                  "start_time": str(int(start_time.timestamp())*1000)
                 }

        response = requests.get(url, params=params, headers=headers)
        if response.status_code == 200:
            # print(response.json())
            data.extend(response.json()['data'])
        else:
            print("Error occurred while fetching data:", response.text)

        # Move the start time to the next interval
        start_time += timedelta(hours=5000)

    return(pd.DataFrame(data))
```

```python
def download_symbol(url: str, start_time: datetime, symbol: str, interval: str) -> pd.DataFrame:
    
    data = []

    headers = {
               "accept": "application/json",
               "coinglassSecret": "91468f8a11bc4f2587cb250789da75dd"
              }
    
    while start_time <= datetime.now():
        
        # Set the parameters for the API request
        params = {
                  "symbol": symbol,
                  "interval": interval,
                  "limit": 5000,
                  "start_time": str(int(start_time.timestamp())*1000)
                 }

        response = requests.get(url, params=params, headers=headers)
        if response.status_code == 200:
            data.extend(response.json()['data'])
        else:
            print("Error occurred while fetching data:", response.text)

        # Move the start time to the next interval
        start_time += timedelta(hours=5000)

    return(pd.DataFrame(data))
```

```python
url = "https://open-api.coinglass.com/public/v2/indicator/open_interest_ohlc"

# Set the time frame for the historical data
start_time = datetime(2022, 11, 29)  # Replace with your desired start time

df = download_pair(url, start_time, "Binance", "BTCUSDT", "h1")
# Drop rows with any NaN value
df = df.dropna()
df = df.reset_index(drop=True)
print(df)
```

```python
# create subplots with two y-axes
fig = make_subplots(specs=[[{"secondary_y": True}]])

# add the trace for the close data to the first y-axis
fig.add_trace(go.Candlestick(
    x=pd.to_datetime(df['t'], unit='ms'),
    open=df['o'],
    high=df['h'],
    low=df['l'],
    close=df['c'],
    name='Close')
)
```
